# 월별 매출액 추세

```python
temp_var = sales_data
temp_var["date"] = pd.to_datetime(sales_data["결제일시"], format="%Y-%m-%d %H:%M")
temp_var["month"] = temp_var["date"].dt.month
temp_var["결제금액"] = temp_var["결제금액"].str.replace(pat=r',', repl=r'', regex=True).apply(lambda x: int(x))
month_acd_pay = temp_var.groupby("month")["결제금액"].sum()

plt.figure(figsize=(10,7))
plt.bar(x=month_acd_pay.index, height=month_acd_pay.values)
plt.xticks(month_acd_pay.index)
plt.xlabel("month")
plt.ylabel("payment")
plt.title("payment by month")
plt.show()
```

# 내가 정의한 문제 - 시간대별 최대 매출 매뉴

# sales_data 가져오기

```python
temp = sales_data
temp["hour"] = temp["date"].dt.hour
temp.info()
temp
```

# '시간 구분' 칼럼 추가 시간대 별로 심야, 오전, 오후, 저녁구분

```python
time = []
for i in temp["hour"]:
    if i >= 1 and i < 8:
        time.append("심야")
    elif i >= 8 and i < 12:
        time.append("오전")
    elif i >= 12 and i < 18:
        time.append("오후")
    else:
        time.append("저녁")
temp["시간구분"] = time
temp
```

# 시간 구분별로 시리즈 가져오기

```python
time = ["오전", "오후", "저녁", "심야"]
ranking = []
for i in time:
    ranking.append( temp[temp["시간구분"] == i].groupby("상품명")["결제금액"].sum())
ranking[0]
```

# 각 시간대별 최대 매출매뉴 시각화

```python
fig, ax = plt.subplots(2, 2, figsize=(15,5), constrained_layout=True)

for i in range(3):
    ranking[i] = ranking[i].sort_values(ascending=False)
    x = np.arange(5)
    y = ranking[i].values[:5]
    ax[i//2][i%2].bar(x, y)
    ax[i//2][i%2].set_xticklabels(ranking[i].index[:6])
    ax[i//2][i%2].title.set_text("{}시간 최대 매출 메뉴".format(time[i]))

plt.show()
```
